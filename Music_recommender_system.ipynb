```python
# Removing outliners

def remove_outliers_zscore(data, z_threshold=3):
    z_scores = (data - data.mean()) / data.std()
    return data[abs(z_scores) <= z_threshold]

numeric_columns = df.select_dtypes(include=['number']).columns

for column in numeric_columns:
    df[column] = remove_outliers_zscore(df[column])

processed_file_path = 'processed_file_final.csv'
df.to_csv(processed_file_path, index=False)
```

```python

from scipy.stats import zscore

file_path = 'processed_file.csv'  
df = pd.read_csv(file_path)

# Define a function to replace outliers with the mode of the column
def replace_outliers_with_mode(data, z_threshold=3):
    z_scores = abs(zscore(data))
    outliers = z_scores > z_threshold
    mode_value = data.mode()[0]
    data[outliers] = mode_value
    return data

# Specify the numeric columns you want to process
numeric_columns = df.select_dtypes(include=['number']).columns

# Replace outliers with mode in the specified numeric columns
for column in numeric_columns:
    df[column] = replace_outliers_with_mode(df[column])

# Save the DataFrame with outliers replaced to a new CSV file
processed_file_path = 'processed_file_fianl1.csv' 
df.to_csv(processed_file_path, index=False)
```

# Music Recommder System

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
song_vectorizer = CountVectorizer()
song_vectorizer.fit(song_library['playlist_genre'])
def song_recommender(song_name, song_library):
    try:
      
        num_cols = ['duration_ms', 'track_popularity', 'danceability', 'energy', 'key', 'loudness',
                    'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

      
        text_vec1 = song_vectorizer.transform(song_library[song_library['track_name'] == song_name]['playlist_genre'])      
        num_vec1 = song_library[song_library['track_name'] == song_name][num_cols].values
        sim_scores = []
        for _, row in song_library.iterrows():
            name = row['track_name']
            text_vec2 = song_vectorizer.transform(song_library[song_library['track_name'] == name]['playlist_genre'])
            
            num_vec2 = song_library[song_library['track_name'] == name][num_cols].values
            
            text_sim = cosine_similarity(text_vec1, text_vec2)[0][0]
            
            num_sim = cosine_similarity(num_vec1, num_vec2)[0][0]
           
            sim = (text_sim + num_sim) / 2
            sim_scores.append(sim)
        
        song_library_copy = song_library.copy()
        song_library_copy['similarity'] = sim_scores
        
        song_library_copy.sort_values(by=['similarity', 'track_popularity', 'track_album_release_date'],
                                      ascending=[False, False, False], inplace=True)
        
        recommended_songs = song_library_copy[song_library_copy['track_name'] != song_name][['track_name', 'track_artist', 'track_album_release_date']][:5]
        return recommended_songs
    except KeyError:
        
        print('{} not found in songs library.'.format(song_name))

user_input = input("Enter a song name: ")

recommended = song_recommender(user_input, song_library)
if recommended is not None:
    print("\nRecommended songs:")
    print(recommended)
```
